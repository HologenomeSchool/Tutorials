### There are some rules for naming variables

- A variable name cannot be a <strong>Python keyword</strong>, like print.
- A variable name cannot contain <strong>spaces</strong>
- The first character must be a <strong>letter</strong> (or an underscore).
- Variables names are <strong>case sensitive</strong>, which means that a variable named car is different than a variable named Car.

### Printing multiple items

<p>Python allow you to display multiple items with a single call to the print function.</p>
- There are different ways to do that, but the best one is using the f-format. It's quite easy.

## Data Types

<p>Python, as expected from a programming language, has several datatypes:</p>



- Text: Strings, or sequences of characters (including whitespaces).
- Numeric: Numbers (integers), numbers with float-point position (floats) and complex numbers.
- Sequence: Contains lists and tupples, which can be viewed as an organized sequence of variables.
- Mapping: Dictionaries, they are also used to organize stuff, but they are not sequential!
- Sets: Sets are used to do group operations with lists and dictionaries, such as intersection, union, bla bla bla.
- Boolean: Boolean are binary variables, they can only produce two results, true or false.

We will not cover sets and tupples as they are not important for this course, but please feel free to check on other sources as they are quite useful!. 

<p>Python doesn't require you to declare the type of your variable, the interpreter will decide the type of your variable based on the value assigned to it.</p>

```python
#Other languages like C++ and java requires you to declare the variable before creating it, something like:

int someNumber;
someNumber = 10
```

```python
#Now Python:
someNumber = 10
print(someNumber)
someNumber = 10.0
print(someNumber)
someNumber = "10"
print(someNumber)
```

<p>Python literally doesn't care, the interpreter will decide the data type on its own. Another example, let's try to confuse the interpreter:</p>

```python
X = "I'm an integer"
print(X)
print(type(X)) #We chained another function here called type, to return the type of the variable
```

<p>It will define the variable type behind the scenes like this:</p>



<p></p>

<h3>Boolean Data type</h3>

<p>Boolean data is a special data type that can have only two different outcomes, True or False, being essentially Binary</p>

<p>A Boolean variable can be used to test the identity of something, or to test if something is inside something else (Like a character inside a prhase).</p>
<p>Because of that, a boolean variable will normaly be created as an answer to a question your code will make to the interpreter using the comparison operator "==".</p>
<strong>Example:</strong>
<p></p>

```python
name = "Bruno"
pi = 3.1415
isEarthFlat = "No"

print(name == "Bruno")
print (pi >= 3.14)
print(isEarthFlat == "Yes")
```

<h3>Operators:</h3>

<p>Operators are speciall symbols that carry out arithmetic or logical operations, and they are separate in the following types:</p>


<p></p>

<h4>Arithmetic operators</h4>
<p>The first we have arithmetic operators. They are used to perform common mathematical operations, like:</p>



```python
a = 3
b = 2
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a**b)
```

<p></p>
<h4>Assignment operators</h4>
<p>Next are the assignment operators, we saw two of them already. They are used to assign values to variables or other data structures (more to it later)</p>



```python
x = 5
x += 3
print(x)
```

<p></p>
<h4>Comparison Operators</h4>
<p>Comparison operators are used to compare two values (It will return a boolean value)</p>



```python
x = 5
print(x==2)
print(x!=2)
print(x>2)
print(x<5)
print(x>=5)
print(x<=5)
```

<p></p>
<h4>Logical Operators</h4>
<p>They are used to combine conditional statements (You can chain a lot of different statements, reducing the size of your code).</p>



```python
x = 5
print(x > 3 and x < 10)
print(x > 3 or x < 4)
print(not(x < 5 and x < 10))
```

<h4>Membership</h4>
<p>These operators are used to test if a sequence is found in something, be a string, object or other data structures.</p>



```python
phrase = "Python is awesome"

print("awesome" in phrase)
print("awesome" not in phrase)
```
