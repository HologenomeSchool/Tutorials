<div style="padding-top:0px;padding-bottom:10px;">
    <h2>Main types of plots</h2>
    <h3>plot(x,y)</h3>
    <p>plot(x,y), is the very same we just used, but we can do more complicated plots with it instead of a single line, like using the following polynomial function:</p>  
</div>

```python
import numpy as np #A library for numerical stuff

x = np.linspace(0,10,100)
y = 4 + 2 * np.sin(2*x)

plt.figure(figsize=[10,4])
plt.plot(x,y, linewidth=2.0)
plt.show()
```

<div style="padding-top:10px;padding-bottom:10px;">
    <h3>scatter(x,y)</h3>
    <p>A <strong>scatter plot</strong> is a typle of plot, or mathematical diagram, using cartesian coordinates to display values for typically two variables for a set of data. It can be used to displays the relationship between two sets of data!</p>  
</div>

```python
#Let's make some random data

np.random.seed(3)
x = 4 + np.random.normal (0,2,24) #We are creating random datapoints but they will follow a normal distribution
y = 4 + np.random.normal (0,2,len(x))

plt.scatter(x, y)

plt.show()
```

<div style="padding-top:10px;padding-bottom:10px;">
    <h3>bar(x,y)</h3>
    <p>A <strong>barplot</strong> or <strong>barchart</strong> is a plot that represents categories as rectangular bars with lengths and heights proportional to the values which they represent.</p>
    <p>A barplot can be plotted horizontally (hbar()) and vertically (bar()).</p>
    <p>They describe the comparisons between discreet categories as one of the axis will represent these categories and the other, their magnitude.</p>
    <p>let's create some bars:</p>
</div>

```python
#Creating the dataset

data = {"C":20, "C++":15, "Java":30, "Python":35}

courses = list(data.keys()) #We are decompressing the keys to a list
values = list(data.values())

plt.figure(figsize = (5,3))

#Creating a barplot
plt.bar(courses, values, color = "blue", width = 0.4)
plt.xlabel("Courses offered")
plt.ylabel("No of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

<div style="padding-top:10px;padding-bottom:10px;">
    <h4>Using multiple arguments</h4>
    <p>You can pass more than one argument by separating the arguments by comma.</p>
    <p>The number of arguments have to match the number of parameters and they will be assigned in an ordered way.</p>
    <p>The first parameter will receive the value of the first argument, the second parameter of the second argument, and so on.</p>
    <p>Let's make use of it to create a function to calculate the area of a rectangle:</p>
</div>

```python
def calculateArea(recLength, recWidth):
    print(f"The area of the rectangle is {recLength*recWidth}.")

calculateArea(5, 10)
```

<div style="padding-top:10px;padding-bottom:10px;">
    <h3>Stacked barplots:</h3>
    <p>As you obviously know by now, as aspiring scientists, is that we can stack barplots to make a... stacked barplot.</p>
    <p>Such graph is used to break down and compare different discreet categories, or parts of the whole.</p>
    <p>Each bar in a standard barchart is divided into a number of sub-bars and stacked on top of each other. Sadly, it's quite complicated to build a stacked barplot using matplotlib alone, as you will see:</p>
</div>

```python
#Creating the data

labels = ["G1","G2","G3","G4","G5"]
men_mean = [20,35,30,35,27]
women_mean = [25,32,34,20,25]
men_sd=[2,3,4,1,2]
women_sd=[3,5,2,3,3]
width=0.35 # The width of the bars


plt.bar(labels,men_mean, width, yerr=men_sd, label="Men")
plt.bar(labels,women_mean, width, yerr=women_sd, label="Women", bottom=men_mean)
plt.ylabel("Scores")
plt.title("Scores by group and gender")
plt.legend()
```

<div style="padding-top:10px;padding-bottom:10px;">
    <p>Complicated, right? But it can be even more complicated. I present to you the <strong>Grouped barplots</strong></p>
</div>

<div style="padding-top:10px;padding-bottom:10px;">
    <h3>Grouped barplots</h3>
    <p>They behave similarly to a stacked barplot, but the visualisation (and construction), is a little different</p>
</div>

```python
labels = ["G1","G2","G3","G4","G5"]
men_mean = [20,35,30,35,27]
women_mean = [25,32,34,20,25]
men_sd=[2,3,4,1,2]
women_sd=[3,5,2,3,3]
width=0.35 # The width of the bars
location = np.arange(len(labels)) # the label locations

plt.bar(location - width/2, men_mean, width, label = "Men")
plt.bar(location + width/2, women_mean, width, label="Women")

plt.ylabel("Scores")
plt.title("Scores by group and gender")
plt.legend()
```

<div style="padding-top:10px;padding-bottom:10px;">
    <h3>Histograms</h3>
    <p>A histogram is a graphical representation of your data distribution. It will organize it into group of datapoints in user-specific ranges.</p>
    <p>It's similar in appearance to a bar graph and it condenses a data series into a visual easy to interpret. Histograms are quite useful to visually inspect the data distribution, including it's normality.</p>
</div>

```python
data = np.random.normal(0,2,250)

#Plotting the histogram.

plt.hist(data, bins = 25, density = True, alpha = 0.6, color = "blue")
plt.show()
```

<p>Yes, it's normally distributed with a nice bell shape, but we knew that already since we created it using a random.normal() method.</p>

<div style="padding-top:10px;padding-bottom:10px;">
    <h3>Boxplots</h3>
    <p>A boxplot is another visual representation of your data as it will give you a good indication of how the values in the data are spread out.</p>
    <p>Although boxplots may seem more primitive in comparison to a histogram, they have the advantage of taking up less space, which is useful when comparing distributions between many groups of datasets.</p>
</div>

```python
np.random.seed(10)
D = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))

fig, ax = plt.subplots()
VP = ax.boxplot(D, positions=[2, 4, 6], widths=1.5, patch_artist=True,
                showmeans=False, showfliers=False,
                medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C0", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C0", "linewidth": 1.5},
                capprops={"color": "C0", "linewidth": 1.5})

plt.show()
```

<div style="padding-top:10px;padding-bottom:10px;">
    <h3>Violin plots</h3>
    <p>It's similar to boxplots, but they also show the probability density of the data at different values.</p>
</div>

```python
np.random.seed(10)
D = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))

fig, ax = plt.subplots()

vp = ax.violinplot(D, [2,4,6], widths = 2,
                  showmeans = True)
plt.show()
```

<div style="padding-top:10px;padding-bottom:10px;">
    <h3>Pipe charts</h3>
    <p>Even though I'm not a huge fan, piecharts have their value. They behave similary to a stacked barplot but are useless when you are comparing diferent samples. However, it's ok to use when you are visualising the data from a single sample.</p>
</div>

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries']
sizes = [20, 15, 30, 10, 25]  # Percentage values for each fruit
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']  # Custom colors
explode = (0.1, 0, 0, 0, 0)  # "explode" the 1st slice (i.e. 'Apples')

# Plot
plt.figure(figsize=(5, 5))  # Set figure size
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')  

plt.title('Fruit Pie Chart')
plt.show()
```

<div style="padding-top:10px;padding-bottom:10px;">
    <h3>Heatmaps</h3>
    <p>Sometimes you need to display data that is dependent on two independent variables as a colour coded image plot.</p>
    <p>Sometimes you just want to make more sense of your boring table.</p>
    <p>Heatmaps are useful for both situations:</p>
</div>

```python
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

## Lets define a 2D array with data that defines the color code. 
## the number of elements in those lists need to match the data along the respective axes
harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])

fig, ax = plt.subplots(figsize = [8,5])
im = ax.imshow(harvest)

ax.set_xticks(np.arange(len(farmers)), labels=farmers)
ax.set_yticks(np.arange(len(vegetables)), labels=vegetables)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

for i in range(len(vegetables)):
    for j in range (len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")

fig.tight_layout()
plt.show()

import pandas as pd
pd.DataFrame(harvest, columns = farmers, index = vegetables)
```

<div style="padding-top:10px;padding-bottom:10px;">
    <p>As you can see, this heatmap is just a visual representation of the dataframe, with the color intensity changing according to the value inside the data cell. It's pretty useful because we are visual animals, it's easier for us to understand visual patterns than... numbers.</p>
</div>

</div>

<div style="padding-top:10px;padding-bottom:10px;">
    <p>We can make use of some attributes to customise our plots. As an example, matplotlib's main plot function accepts as input an array of coordinates, but also, a string abbreviation indicating colour and line style</p>
    <p>Let's say we want to change the line to dashes and the colour to blue:</p>
</div>

```python
data = list(range(10))

plt.plot(data, "b--")
```

<div style="padding-top:10px;padding-bottom:10px;">
    <p>Specifying both colour and line style in one go is convenient, but not optimal.</p>
    <p>If you want finer control you might prefer to break it down to attributes, like <strong>color</strong> and <strong>linestyle</strong>:</p>
</div>

```python
plt.plot(data, color="blue", linestyle="--")
```

<div style="padding-top:10px;padding-bottom:10px;">
    <h3>The color attribute:</h3>
    <p>Although this notation is handful, and will almost certainly cover your needs, the limit of 170 colours is in fact a major limitation.</p>
    <p>To overcome it we can use RGB, in which you use a tuple containing the amount of red, green and blue (and alpha transparency for RGBA). With that we can achieve up to 16 million colours.</p>
    <figure>
        <center><img src="Images/rgb_model.png"></center>
        <center><caption>All colours are made of these 3 basic ones</caption></center>
    </figure>
    <p>Another way is to use the hex notation, a hexadecimal representation of the RGB system:</p>
    <figure>
        <center><img src="Images/colourTable.png" style="width:20em;"></center>
    </figure>
    <p>There is a catch though, matplotlib will only accept RGB in a range from 0 to 1, so you will have to transform the numbers.</p>
    <p>To retrieve the hex and RGB codes for a colour that you might use, you can use online services such as <a href="https://imagecolorpicker.com/">imagecolorpicker</a>.</p>
    <p>As an example, let's plot the same plot using the 3 notations:</p>
</div>

```python
plt.plot(data, color = "skyblue")
```

```python
plt.plot(data, color = "#56B4E9")
```

```python
plt.plot(data, color = (0.31,0.7,0.91))
```

<div style="padding-top:10px;padding-bottom:10px;">
    <h3>Different colours for different elements of the plot</h3>
    <p>Some plot functions accepts multiple colours for different elements in a plot, such as stacked bar plots or slices of a pie chart! To do so, you will pass the colours as a list to the colors attribute:</p>

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries']
sizes = [20, 15, 30, 10, 25]  # Percentage values for each fruit
explode = (0.1, 0, 0, 0, 0)  # "explode" the 1st slice (i.e. 'Apples')

# Plot
plt.figure(figsize=(5, 5))  # Set figure size
plt.pie(sizes, explode=explode, labels=labels, colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink'], autopct='%1.1f%%', shadow=True, startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')  

plt.title('Fruit Pie Chart')
plt.show()
```

<div style="padding-top:10px;padding-bottom:10px;">
    <h3>Saving a figure object with the savefig() method.</h3>
    <p>You can easily save the figure object containing your plots by using the savefig method:</p>
</div>

```python
plt.plot(data, color = (0.31,0.7,0.91))
plt.savefig("lineplot.png", dpi=150)
```

<div style="padding-top:10px;padding-bottom:10px;">
    <p>The first attribute is the pathway and name of your plot, followed by the suffix (here is png). The second attribute is the figure quality measured in dpi (dots per inches).</p>
    <p>Dpi is only valid for raster graphics format and not for vectorized figures.</p>
    <p><center><span style = "color:red; font-weight:bold;">But what's the difference between them?</span></center></p>
</div>

<div style="padding-top:10px;padding-bottom:10px;">
    <p>A raster graphic is a figure composed by tiny squares, also known as pixels.</p>
    <p><center><img src="Images/fish.png"></center></p>
    <p>higher valules of DPI means higher number of pixels, which is translated in figures with better quality.</p>
    <p>A vector, on the other hand is not truly a figure but a mathematical model. Such model is interpreted by the computer and visualised as a figure.</p>
    <p>And because of this property, a vectorized figure will <strong>Never lose quality</strong>, It doesn't matter how much you zoom in.</p>
    <p> </p>
    <p><center><img src="Images/vector.png"></center></p>
</div>
