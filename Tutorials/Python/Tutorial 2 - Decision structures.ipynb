<h3>Example:</h3>
<p>Let's create a piece of code to reward an employee with a bonus of &#8364;500 if their sales exceed &#8364;5000.</p>

```python
sales = 4999
bonus = 0 #We need to initialize the bonus as 0 before the if clause to avoid problems

if sales> 5000:
    bonus = 500

print(f"Bonus value was {bonus}")
```

 <h2>The if/else structure:</h2>

<p>The if/else structure is also known as the <strong>Dual alternative decision structure</strong></p>
<p>Which means that we can do stuff if a condition is true, and another stuff if the condition is false!</p>

<p>the syntax is:</p>
<div style="text-align: center;"><p><strong>if</strong> condition: <br>
    <span style="padding:20px;">statement(s)</span><br>
    <strong>else:</strong><br>
    <span style="padding:20px;">other statement(s)</span>
</p></div>

<p>With it we can solve the flawed husband's logic</p>


<center><img src="Images/husband2.png"></center>

<p>And that's it!</p>
<p>Let's try to use it to make that bonus more fair...</p>

```python
sales = 4999
bonus = 0 #We need to initialize the bonus as 0 before the if clause to avoid problems

if sales> 5000:
    bonus = 500
else:
    bonus = 100

print(f"Bonus value was {bonus}")
```

<p>It still doesn't seem fair, if you sold 0 to 4999, you would be entitled </p>

<p>The answer to that is to create a <strong>nested</strong> decision structure.</p>

<p>It works like this:</p>

<center><img src = "Images/Nested.png"></center>

<h3>What is going on?:</h3>
<p>The first if clause will be tested, if its false, then the program will stop, but if it's true, it will test the second if clause.</p>
<p><strong>Let's see how to do it programmatically:</strong></p>

```python
sales = 4999
bonus = 0 #We need to initialize the bonus as 0 before the if clause to avoid problems

if sales> 1000:
    if sales > 5000:
        bonus = 500
    else:
        bonus = 100
else:
    bonus = 0

print(f"Bonus value was {bonus}")
```

<p>It looks busy, and it is. Nesting is always ugly, and there is a better way to do this.</p>
<p>The next and final decision structure is known as the <strong>if-elif-else</strong> structure where elif stands for "else if" and it removes the need of nesting anything.</p>

<p>the syntax is:</p>
<div style="text-align: center;"><p><strong>if</strong> condition: <br>
    <span style="padding:20px;">statement 1</span><br>
    <strong>elif:</strong><br>
    <span style="padding:20px;">statement 2</span><br>
    <strong>else:</strong><br>
    <span style="padding:20px;">statement 3</span>
</p></div>

<p>Let's re-write the previous example for you to understand why this structure is more optimal:</p>

```python
sales = 4999
bonus = 0 #We need to initialize the bonus as 0 before the if clause to avoid problems

if sales > 1000:
    bonus = 100
elif sales >=5000:
    bonus = 500
else:
    bonus = 0

print(f"Bonus value was {bonus}")
```

<h3>The While loop</h3>
<p></p>

<p><strong>We can easily retrieve information from this structure:</strong></p>

<div style="display:flex;padding-top:10px;padding-bottom:10px;">
<div>
    <p><strong>The while loop</strong> is also known as the Condition Controlled loop. Meaning that While a condition is true, the loop will redo the process until the condition is not true anymore</p>
    <p><strong>The while loop has two parts (Just like decision structures)</strong></p>
    <ul>
        <li><strong>The condition</strong> tested for <strong style="color:blue">true</strong> or <strong style="color:red">false</strong></li>
        <li>The statement<strong></strong> that will be repeated while the condition is true.</li>
    </ul>
    <p>the syntax is:</p>
<p><strong>while</strong> condition: <br>
    <span style="padding:20px;">statemens</span><br>
    
</p>
</div>
    <div>
        <img src="Images/while.png" style="width:40em">
    </div>
            
</div>

<p><strong>Let's check one example:</strong></p>

```python
num = 5

while num <10:
    print(num)
    num+=1
```

<div style="padding-top:10px;padding-bottom:10px;">
    <p>What happened here was that the loop will loop while the variable num is higher than <strong>10</strong>.</p>
    <p>In each loop, num will be incremented by one, and when num is higher than <strong>10</strong>, the loop will be finished. The num variable is what is known as a counter, it will count the number of times the loop will loop and will be used as a way out.</p>
    <p>But what happens if your condition doesn't change? Well...</p>
    <p><center><img src="Images/infiniteloop.png"></center></p>
    <p>I will show you an example, what would happen if I remove the counter (<strong>If you execute the cell, please press the stop button on the top bar.</strong>):</p>
</div>

```python
num = 5

while num <10:
    print(num)
```

<p>There is a better way to work with for loops, let me introduce you to the <strong>range function</strong></p>

<div style="padding-top:10px;padding-bottom:10px;">
<h3>Using the range function for loops:</h3>
    <p>The The range function simplifies the process of writing a <strong>for</strong> loop</p>
    <p>You can think of the range function as a way to generate a list of things, which is generally used to iterate over with loops, let me re-do the first example: </p>
</div>

```python
for num in [0,1,2,3,4]:
    print(num)
```

<p>now with the range function</p>

```python
for num in range(5):
    print(num)
```

<div style="padding-top:10px;padding-bottom:10px;">
    <p>The range function can be used like this, with a single number indicating where to stop the loop, but you can do much more with it. The full syntax for the range funcion is</p>
    <p><center>range([start,] stop [,step])</center></p>
    <ul>
        <li><strong>start:</strong> Will define the starting position for the list, sometimes we just don't want it to start at 0.</li>
        <li><strong>stop:</strong> It's the most important one and mandatory, will define the ending position.</li>
        <li><strong>step:</strong> It's literally the step, if we want to create a list jumping one or two numbers at a time.</li>
    </ul>
    <p>The thing is, if we use the range function with a single arguments, it will be the <strong>stop</strong> position. If we use two arguments, it will be the <strong>start</strong> and <strong>stop</strong>. If we use 3, the last attribute will be the <strong>step</strong></p>
    <p>Let's try it out:</p>
</div>

```python
for number in range(1,9):
    square = number**2
    print(f"Number:{number} \t Square:{square}")
```

<p>Or:</p>

```python
for num in range(1,10,2):
    print(num)
```

<div style="padding-top:10px;padding-bottom:10px;">
    <h2>Nested loops</h2>
    <p>Just like in decision structures, a nested loop is a loop inside another loop</p>
    <p>The key points about nested loops are:</p>
    <ul>
        <li>The inner loop most go through all iterations for each iteration of the outer loop, just like an analog clock!!</li>
    </ul>
    <p><center><img src="Images/clock.png" style="width:200px;"></center></p>
    <p>So, for every time the second hand loops, the minutes hand will move one position, and for every time the minutes hand loops, the hour hand will move one position.</p>
    <p>That's exactly how a nested loop works and... you can easily get yourself in an infinite loop! Well, let's have some examples:</p>
</div>

```python
for num1 in range(3):
    print("The outer loop moved one position")
    print(num1)
    for num2 in range(2):
        print("The inner loop moved one position")
        print(num2)
        print("The inner loop finished")
    print("The outer loop finished")
```

<div style="padding-top:10px;padding-bottom:10px;">
    <p>We can even build a clock! (Well, sort of)</p>
</div>

```python
for hours in range(24):
    for minutes in range(60):
        for seconds in range(60):
            print(f"Hours:{hours}, minutes:{minutes}, seconds:{seconds}")
```
